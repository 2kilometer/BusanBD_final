```python
def getDF(data, cols_dict) :
    rows = []
    for i in range(len(data)) :
        row_data = {}
        for k,v in cols_dict.items() :
            row_data[k] = data[i].find_all(k)[0].text
        rows.append(row_data)
    df = pd.DataFrame(rows)
    return(rows)
```

```python
data_total = []
for idx in range(1, 34500, 1000):
    print(idx)
    
    apikey = "1a55d53d9b3440a1874d"
    url = f"http://openapi.foodsafetykorea.go.kr/api/{apikey}/I0030/xml/{idx}/{idx+1000}"
    rs_data = requests.get(url).text
    rs_xml = BeautifulSoup(rs_data, "lxml")
    
    data_total += getDF(rs_xml.find_all("row"), cols_dict)
    
df = pd.DataFrame(data_total)
df
```

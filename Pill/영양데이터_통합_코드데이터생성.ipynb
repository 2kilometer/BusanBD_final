```python
### 성분명 파일 기준으로 전처리하기
# - 매개변수(nt)에 성분명 입력

def setProdData(nt):
    # 성분별 파일 불러오기
    df = pd.read_csv(f"./{nt}.csv", usecols=['prod_gb', 'prod_mnf', 'prod_nm', 'rv_scr', 'rv_cnt', 'prod_hff', 'main_func', 'sub_func', 'intk_time', 'intk_how'])
    # 태그 안 데이터 추출
    df["prod_nm"] = df["prod_nm"].str.extract(r'">(.*?)</')
    df["prod_mnf"] = df["prod_mnf"].str.extract(r'">(.*?)</')
    df["rv_scr"] = df["rv_scr"].str.extract(r'">(.*?\d+\.\d+)</')
    df["rv_cnt"] = df["rv_cnt"].str.extract(r'">(.*?\d+)개의 리뷰</')
    df["main_func"] = df["main_func"].str.findall(r'">(.*?)</')
    df["intk_how"] = df["intk_how"].str.extract(r'">(.*?)</')

    # uniq_func을 사용해 주요기능 82->28개로 축소분류
    for i in range(len(df)):
      df["main_func"][i] = [uniq_func[v] for v in df["main_func"][i]]

    # "섭취 정보" 컬럼 생성
    # --- 섭취 시간 + 섭취 방법을 합친 컬럼
    for i in range(len(df)):
        intk_time = ""
        for v in eval(df["intk_time"][i]):
            if intk_time : intk_time += ("/" + v.split("|")[0])
            else : intk_time += v.split("|")[0]
        df["intk_time"][i] = intk_time
    df["intk_info"] = "*"+df["intk_time"]+"*   "+df["intk_how"]

    ###### 최종 데이터프레임 생성 ######
    df = df[["prod_gb", "prod_nm","prod_mnf","rv_scr","rv_cnt","main_func","intk_info"]]

    return df
```

# 통합
---

* **pill_cd 코드데이터 미포함**


```python
total = pd.concat([setProdData(nt) for nt in nts.columns[1:]],ignore_index=True)
total.columns = ['nutr_nm', 'pill_nm', 'pill_mnf', 'pill_rv', 'pill_rvnm', 'pill_mainfunc', 'pill_info']
total
```

* **pill_cd 코드데이터 포함**

```python
nutr = pd.read_csv("./nutr.csv")
nutr_tag = pd.read_csv("./nutr_tag.csv", encoding="euc-kr")
func = pd.read_csv("./func.csv", dtype=str)
```

```python
func["func_nm"] = func["func_nm"].str.replace("&", "&amp;")
func
```

* **제품 코드 생성**

```python
# pill_cd 코드 생성 후 데이터프레임으로 열 추가
pill_cd = pd.DataFrame()
pill_cd = pd.merge(left = total, right = nutr, how = "left")
cd = pd.DataFrame({"pill_cd":list(map(lambda x: str(x).zfill(6), range(1, len(pill_cd)+1)))})
pill_cd = pd.concat([pill_cd, cd], axis=1)
pill_cd["pill_cd"] = pill_cd["nutr_cd"]+pill_cd["pill_cd"]

pill_cd
```

* **제품 코드 저장**

```python
temp = pill_cd[['pill_cd','pill_nm', 'pill_mnf', 'pill_rv', 'pill_rvnm', 'pill_mainfunc', 'pill_info']]
temp
temp.to_csv("pill.csv", index=False)
```

* **통합 코드 생성**

```python
# pill_cd 코드 생성 후 데이터프레임으로 열 추가
total_cd = pd.DataFrame()
total_cd = pd.merge(left = total, right = nutr, how = "left")
cd = pd.DataFrame({"pill_cd":list(map(lambda x: str(x).zfill(6), range(1, len(total_cd)+1)))})
total_cd = pd.concat([total_cd, cd], axis=1)
total_cd["total_cd"] = total_cd["nutr_cd"]+total_cd["pill_cd"]

total_cd = pd.merge(left = total_cd, right = nutr_tag, how = "left", left_on="nutr_tag", right_on="tag_nm").explode("pill_mainfunc").reset_index()
total_cd = pd.merge(left = total_cd, right = func, how = "left", left_on="pill_mainfunc", right_on="func_nm")

total_cd
```
